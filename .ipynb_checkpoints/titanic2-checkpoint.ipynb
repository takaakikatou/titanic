```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
train_data = train_data.fillna(train_data.mean())
```

```python
train_data = train_data.drop('Cabin', axis = 1)
```

```python
train_data = train_data.dropna()
```

```python
test_data = test_data.fillna(test_data.mean())
```

```python
test_data = test_data.drop('Cabin', axis = 1)
```

```python
plt.figure(figsize = (15,20))
```

```python
sns.heatmap(train_data.corr(), vmin = -1.0, vmax = 1.0, annot = True, cmap = 'Reds', linewidths = 0.1)
```

```python
train_data_1 = pd.get_dummies(train_data['Embarked'])
```

```python
train_data = train_data.drop('Ticket', axis = 1)
```

```python
train_data = train_data.drop('Name', axis = 1)
```

```python
train_data['Sex'] = train_data['Sex'].apply(lambda x:0 if x == 'female' else 1)
```

```python
train_data_1 = pd.get_dummies(train_data)
```

```python
X = train_data_1.drop('Survived', axis = 1)
```

```python
y = train_data_1['Survived']
```

```python
lr = LogisticRegression(max_iter=10000)
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.datasets import load_breast_cancer
```

```python
score = cross_val_score(lr, X, y, cv=3)
```

```python
print("{:.4f}".format(np.mean(score)))
```

```python
lr.fit(X,y)
```

```python
print("テストデータの評価")
print("{:.4f}".format(lr.score(X, y)))
```
